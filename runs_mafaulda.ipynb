```python
df_normal = pd.read_csv(cur_path + "/base/normal.csv",sep=";")
df_hm = pd.read_csv(cur_path + "/base/horizontal-misalignment.csv",sep=";")
df_vm = pd.read_csv(cur_path + "/base/vertical-misalignment.csv",sep=";")
df_imbalance = pd.read_csv(cur_path + "/base/imbalance.csv",sep=";")
df_over = pd.read_csv(cur_path + "/base/overhang.csv",sep=";")
df_under = pd.read_csv(cur_path + "/base/underhang.csv",sep=";")
```

```python
df_normal.shape, df_hm.shape, df_vm.shape, df_imbalance.shape, df_over.shape, df_under.shape
```

```python
df_normal['target'] = 0
df_hm['target'] = 1
df_vm['target'] = 1
df_imbalance['target'] = 1
df_over['target'] = 1
df_under['target'] = 1
```

```python
df_normal['target2'] = 'normal'
df_hm['target2'] = 'horizontal-misalignment'
df_vm['target2'] = 'vertical-misalignment'
df_imbalance['target2'] = 'imbalance'
df_over['target2'] = 'overhang'
df_under['target2'] = 'underhang'
```

```python
df = pd.concat([df_normal, df_hm, df_vm, df_imbalance, df_over, df_under])

y = df['target2']
X = df.drop('target2', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

X_train['target2'] = y_train
X_test['target2'] = y_test

X_train.to_csv(cur_path + '/runs/1/train.csv', index=False, sep=";")
X_train.to_csv(cur_path + '/runs/1/test.csv', index=False, sep=";")
```
