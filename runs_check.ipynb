```python
import pandas as pd
import numpy as np
import os
```

```python
from google.colab import drive
drive.mount('/content/drive')
cur_path = "drive/MyDrive/Mafaulda_organizado"
```

```python
experiments_dir = cur_path + '/runs'
```

```python
experiments = next(os.walk(experiments_dir))[1]
```

```python
file_path = cur_path + '/results/report.txt'
```

```python
for experiment in experiments:
  df_train = pd.read_csv(experiments_dir + '/' + experiment + '/' + 'train.csv', sep=";")
  df_test = pd.read_csv(experiments_dir + '/' + experiment + '/' + 'test.csv', sep=";")

  train_value_counts = df_train['target2'].value_counts()
  test_value_counts = df_test['target2'].value_counts()

  with open(file_path, 'a') as file:
    file.write('===== Experiment ' + experiment + '=====\n')
    file.write('Dataset Train\n')
    file.write(str(train_value_counts))
    file.write('\n')
    file.write('Dataset Test\n')
    file.write(str(test_value_counts))
    file.write('\n')
```
